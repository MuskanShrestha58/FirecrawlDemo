# Extracting using LLM provided by Firecrawl

```python
from pydantic import BaseModel, Field
from typing import List
```

```python
from pydantic import BaseModel, Field
from typing import List
from firecrawl import FirecrawlApp

# Define the schema for individual articles
class PageSchema(BaseModel):
    title: str
    description: str
    land_area: str
    location: str
    amneties: list
    media_urls: list
    price: str
    by: str
    URL: str

# Define the schema for top articles, allowing a maximum of 5 articles
class TopPropertiesSchema(BaseModel):
    top: List[PageSchema] = Field(..., max_items=10, description="Top 10 properties")

# Initialize the FirecrawlApp with your API key
app = FirecrawlApp(api_key="fc-485e7f28681a48ec8d102a0e2c543392")

# Define the payload for scraping the URL
data = app.scrape_url('https://basobaas.com/for-sale', {
    'extractorOptions': {
        'extractionSchema': TopPropertiesSchema.model_json_schema(),
        'mode': 'llm-extraction'
    },
    'pageOptions': {
        'onlyMainContent': True
    }
})

# Extract the relevant data
data_extracted = data.get("llm_extraction")

# Print the extracted data
print(data_extracted)
```

```python
# Extraction Output 
# Save the extracted data to a JSON file
output_file = dpath+ "llm_extracted_data.json"
with open(output_file, 'w') as json_file:
    json.dump(data_extracted, json_file, indent=4)

print(f"Content saved to {output_file}")
```
